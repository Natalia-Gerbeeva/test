# Задание 1

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
fl_movies = "movies.csv"
fl_ratings = "ratings.csv"

ds_movies = pd.read_csv(fl_movies)
ds_movies.head()
```

```python
ds_movies.columns = ['movieId', 'title', 'genres']
ds_movies.head()
```

```python
ds_ratings = pd.read_csv(fl_ratings)
ds_ratings.head()
```

```python
dataframes = [ds_movies, ds_ratings]

result_set = pd.concat(dataframes)
result_set.head()

#result_set
```

```python
ds_movies.set_index('movieId').join(ds_ratings.set_index('movieId'))

ds_movies.head()
```

```python
ds_mov_rating = pd.merge(ds_movies, ds_ratings, on='movieId', how='left')

ds_mov_rating.head()
```

```python
def get_score(row):
    return "Низкий рейтинг" if float(row['rating']) <= 2.0 else "Средний рейтинг" if float(row['rating']) <= 4.0 else "Высокий рейтинг" if float(row['rating']) > 4.0 else None
```

```python
ds_mov_rating['score'] = ds_mov_rating.apply(get_score, axis=1)

ds_mov_rating.head()
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
print(geo_data)
```

```python
ds_keywords = pd.read_csv("keywords.csv")
ds_keywords.head()

ds_keywords
```

```python
#идея 1 - сделать сплит строки файла, Но:
#со сплитом плохой вариант т.к. город может состоять из 2х слов, напр "старый оскол"

#надо учесть что в строке могут быть 2 города

def get_region(row):
    result = []
    for region, location in geo_data.items():

        for city in location:

            if city in row["keyword"]:
                result.append(region)  

    result = sorted(set(result))

    return "|".join(result) # if "|".join(result) is not None else "1"
```

```python
ds_keywords['region'] = ds_keywords.apply(get_region, axis=1)

#ds_keywords.head()

#ds_keywords.sort_values(by=['region'], ascending=True) -- Не работает

filtered_ds_keywords = ds_keywords

filtered_ds_keywords[filtered_ds_keywords['region'].map(len) > 0] #-- Не работает

#filtered_ds_keywords.head() 

#filtered_ds_keywords.dropna(thresh=1) -- Не работает

#filtered_ds_keywords = ds_keywords[ds_keywords['region'].notnull()] -- Не работает

#Как правильно очистить датасет от None значений по полю?

filtered_ds_keywords.to_csv("filtered_output.csv", index=False)

filtered_ds_keywords.head() 
#filtered_ds_keywords.region.notnull() -- Не работает

#Как правильно очистить датасет от None значений по полю?
```
