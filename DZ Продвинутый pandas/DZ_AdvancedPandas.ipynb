**Задание 1**
Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:
- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений


```python
visit_log = pd.read_csv('visit_log.csv', delimiter=";")
visit_log[:18]
```

```python
#если источник traffic_source равен yandex или google, то в source_type ставится organic
SOURCE = ['yandex', 'google']

print(visit_log['traffic_source'])
    
visit_log.loc[visit_log.traffic_source.isin(SOURCE), 'traffic_source'] = 'organic'
visit_log[:18]
```

```python
#ля источников paid и email из России - ставим ad
SOURCE_2 = ['paid', 'email']

visit_log.loc[visit_log.traffic_source.isin(SOURCE_2) & visit_log.region.isin(['Russia']), 'traffic_source'] = 'ad'
visit_log[:18]
```

```python
#для источников paid и email не из России - ставим other
visit_log.loc[visit_log.traffic_source.isin(SOURCE_2) & ~visit_log.region.isin(['Russia']), 'traffic_source'] = 'other'
visit_log[:18]
#Ukraine -> Other
```

### Домашнее задание 3
В датафрейме data создайте столбец lemmas, в котором вычислите леммы поисковых запросов из столбца keyword. Леммы должны иметь строковый тип.

```python
from pymystem3 import Mystem
```

```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})

data.head()
```

```python
search = 'курс гривны к рублю рубли рублях'
m = Mystem()
print(m)
lemmas = m.lemmatize(search)
lemmas
```

```python
''.join(lemmas)
```

```python
data['lemmas'] = data['keyword'].apply(lambda search: ''.join(Mystem().lemmatize(search))[:-1])
data[:10]
```
