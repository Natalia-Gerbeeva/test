# Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.



```python
import pandas as pd
```

```python
data = pd.read_csv('ratings.csv')
data.head()
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
data_50 = data[data.rating.eq(5.0)]
data_50.head()
```

```python
res = data_50.groupby('movieId').count().sort_values('timestamp', ascending=False)
```

```python
result = res.join(movies.set_index('movieId'), on='movieId')
result.head()
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.



```python
consumption = pd.read_csv('power.csv')
consumption.head()
```

```python
consumption_result = consumption[consumption.country.isin(['Latvia', 'Lithuania', 'Estonia']) 
                                 & consumption.year.isin([2005, 2006, 2007, 2008, 2009, 2010])
                                 & consumption.category.isin([4, 12, 21])
                                 & consumption.quantity.ge(0.0)
                                ]
consumption_result.head()
```

```python
consumption_result.sum(axis = 0, skipna = True) 
```

```python
print(consumption_result['quantity'].sum())
```
